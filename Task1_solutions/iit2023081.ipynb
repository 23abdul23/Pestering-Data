```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
```

```python
dataset_path = "/kaggle/input/datasetoc"

train_dir = os.path.join(dataset_path, 'dataset_train')
test_dir = os.path.join(dataset_path, 'dataset_train')

# Change this code
```

```python
train_data = ImageDataGenerator(rescale=1.0/255) 
test_data = ImageDataGenerator(rescale=1.0/255)


train_generator = train_data.flow_from_directory(
    train_dir,
    target_size=(224,224),
    batch_size=32,
    class_mode='categorical'    
)

test_generator = test_data.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

print("Classes found in the dataset:", train_generator.class_indices)

# Do not change this code
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Write the code to display one image of all the 22 classes.
class_names = list(train_generator.class_indices.keys())
num_classes = len(class_names)
class_images = {}
```

Run all the cells and the output must be clearly visible for all the cells.

```python
for images, labels in train_generator:
    for i, label_vector in enumerate(labels):
        class_index = np.argmax(label_vector)
        class_name = class_names[class_index]
        
        if class_name not in class_images:
            class_images[class_name] = images[i]
            
        if len(class_images) == num_classes:
            break
    if len(class_images) == num_classes:
        break
```

```python
plt.figure(figsize=(15, 15))
for i, (class_name, image) in enumerate(class_images.items()):
    plt.subplot(5, 5, i + 1) 
    plt.imshow(image)
    plt.title(class_name)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
